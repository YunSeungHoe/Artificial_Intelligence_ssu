```python
#########################################################
# 데이터 가공 코드
# using 
# 1. xlsx/recipe_name.xlsx
# 2. xlsx/recipe_ingredient.xlsx
# 3. xlsx/recipe_process.xlsx 
# result 
# recipe_data.xlsx
#########################################################
import pandas as pd
import openpyxl
from datetime import date

df1 = pd.read_excel("xlsx/recipe_name.xlsx")         # 요리 이름
df2 = pd.read_excel("xlsx/recipe_ingredient.xlsx")   # 요리 재료
df3 = pd.read_excel("xlsx/recipe_process.xlsx")      # 요리 과정

name_set = set(int(df1.values[i][0]) for i in range(len(df1)))
ingr_set = set(int(df2.values[i][0]) for i in range(len(df2)))
proc_set = set(int(df3.values[i][0]) for i in range(len(df3)))

# # 요리명이 더 적다
# print("name len : {} \ningr len : {} \nproc len : {}".format(len(name_set), len(ingr_set), len(proc_set)))
# # name ingr 비교
# print("ingr - name : {} \ningr ^ name : {}".format(len(ingr_set-name_set), len(ingr_set^name_set)))
# print("(ingr - name) ^ (ingr ^ name) : {}".format(len((ingr_set-name_set)^(ingr_set^name_set))))
# # name proc 비교
# print("proc - name : {} \nproc ^ name : {}".format(len(proc_set-name_set), len(proc_set^name_set)))
# print("(proc - name) ^ (proc ^ name) : {}".format(len((proc_set-name_set)^(proc_set^name_set))))
# # 위의 과정을 통해 3개의 xlsx 파일의 요리수를 맞춰줌. 
# # 행의 길이 name(이름) < proc(과정) < ingr(재료)

# while loop 이용해서 데이터 가공
# data[n][0] : 이름
# data[n][2] :  (name) 요리법
#               필요한 재료 : (재료1), (재료2), (재료3), ... (재료n)
#
#               1. (과정1)
#               2. (과정2)
#               ...
#               n. (과정n)
# 요리 코드 : df1.values[n][0] / df2.values[n][0] / df3.values[n][0]
# 요리 이름 : df1.values[n][1]
# 요리 재료 : df2.values[n][2] / df2.values[n][3]
# 요리 과정 : df3.values[n][1] / df3.values[n][2]

wb = openpyxl.Workbook()
wb.active.title = "data"
wb1 = wb["data"]
wb1['A'+str(1)].value = "rule_and"
wb1['B'+str(1)].value = "rule_or"
wb1['C'+str(1)].value = "response"

name_cnt = 0
ingr_cnt = 0
proc_cnt = 0
frame1 = " 요리법\n필요한 재료 : "
frame2 = ", "
frame3 = "\n\n"
frame4 = ". "
response = ""
rule_or = "요리 조리 레시피 어떻게 만들 만드는 방법 법 순서"
while name_cnt <= len(df1):
# while True:
    if name_cnt == len(df1) and ingr_cnt == len(df2) and proc_cnt == len(df3):
        break
    if df1.values[name_cnt][0] == df2.values[ingr_cnt][0] == df3.values[proc_cnt][0]:
        rule_and = df1.values[name_cnt][1]
        response = rule_and + frame1
        while True:
            response += str(df2.values[ingr_cnt][2]) + " " + str(df2.values[ingr_cnt][3])
            ingr_cnt += 1
            if ingr_cnt == len(df2): break
            elif df2.values[ingr_cnt][0] == df1.values[name_cnt][0]:
                response += frame2
            else: break
        response += frame3
        while True:
            response += str(int(df3.values[proc_cnt][1])) + frame4 + str(df3.values[proc_cnt][2])
            proc_cnt += 1
            if proc_cnt == len(df3): break
            elif df3.values[proc_cnt][0] == df1.values[name_cnt][0]:
                response += "\n"
            else: break
        name_cnt += 1
        
    # # rule_and, response 저장 완료 
    # print(rule_and)
    # print(rule_or)
    # print(response)
    # print("---------------------------")   
    
    # 새로운 파일에 저장해야한다. 
    wb1['A'+str(name_cnt+1)].value = rule_and
    wb1['B'+str(name_cnt+1)].value = rule_or
    wb1['C'+str(name_cnt+1)].value = response
wb.save('/home/misys/github/artificial_Intelligence_ssu/Project4/recipe_data.xlsx')
```
